## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
# create new df with only sn gender
gender_df=purchase_data_df.loc[:,['SN','Gender']]

# # drop duplicates with subset sn
gender_df.drop_duplicates('SN',inplace=True)

# group by genders
gender_grouped_df=gender_df.groupby('Gender').count()

sn_count=gender_grouped_df.loc[:, 'SN'].sum()


# create gender demo percents variable that divides counts by sn count
gender_grouped_df.loc[:,'Percentage of Players']= gender_grouped_df.loc[:, 'SN']/sn_count


# create df with renamed columns, formatting for percent
new_columns={
    
    'SN': 'Total Count'
}

gender_demo_df=gender_grouped_df.rename(columns=new_columns)
gender_demo_df.loc[:, 'Percentage of Players']=gender_demo_df.loc[:,'Percentage of Players'].map("{:,.2%}".format)
gender_demo_df.index.name = None
gender_demo_df.sort_values('Percentage of Players', ascending=False)
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Bin the purchase_data data frame by age
manual_bins = [0, 9.9,14.9, 19.9, 24.9, 29.99, 34.99, 39.99, 100]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

purchase_data_df.loc[:, "Age Group"]=pd.cut(purchase_data_df["Age"], bins=manual_bins, labels=group_names, include_lowest=True)

purchase_data_df
```

```python
# create a group based off of age bins
age_df=purchase_data_df.groupby('Age Group')
```

```python
# Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below
age_purchase_cnt=age_df['Purchase ID'].count()

# average purchase price = total price/purchase count
age_cum_price=age_df['Price'].sum()

# cum_purchase=purchase_data_df['Purchase ID'].count()

age_avg_purchase_price=age_cum_price/age_purchase_cnt


# average purchase total per person = total price/age_count

age_avg_pp=age_cum_price/age_count
```

```python
# Create a summary data frame to hold the results
age_purchase_columns={
    "Purchase Count": age_purchase_cnt,
    "Average Purchase Price": age_avg_purchase_price,
    "Total Purchase Value": age_cum_price,
    "Avg Total Purchase per Person": age_avg_pp
}

age_summary_df=pd.DataFrame(age_purchase_columns)
age_summary_df.index.name = "Age Ranges"
age_summary_df
```

```python
# # Optional: give the displayed data cleaner formatting

age_summary_df.style.format("${:,.2f}", subset=['Average Purchase Price', 'Total Purchase Value', 'Avg Total Purchase per Person'])
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Retrieve the Item ID, Item Name, and Item Price columns
items_df=purchase_data_df.loc[:, ['Item ID', 'Item Name', 'Price']]
items_df
```

```python
# Group by Item ID and Item Name. 
grouped_items_df=items_df.groupby(['Item ID', 'Item Name'])
grouped_items_df.count()
# Perform calculations to obtain purchase count, average item price, and total purchase value
purchase_count=grouped_items_df['Item ID'].count()
item_cum_value=grouped_items_df['Price'].sum()
avg_item_price=item_cum_value/purchase_count
```

```python
# Create a summary data frame to hold the results
item_summary_columns={
    "Purchase Count": purchase_count,
    "Average Item Price": avg_item_price,
    "Total Purchase Value": item_cum_value
}

item_summary_df=pd.DataFrame(item_summary_columns)
item_summary_df
```

```python
# Sort the purchase count column in descending order
item_summary_df=item_summary_df.sort_values(by=['Purchase Count'], ascending=False)
item_summary_df
```

```python
# * Optional: give the displayed data cleaner formatting
item_summary_df['Average Item Price']=item_summary_df['Average Item Price'].map("${:,.2f}".format)
item_summary_df['Total Purchase Value']=item_summary_df['Total Purchase Value'].map("${:,.2f}".format)
```
